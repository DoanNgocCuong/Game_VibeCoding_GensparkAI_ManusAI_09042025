## Chuyện giải quyết vấn đề
Trong công việc thì thường là có 3 loại vấn đề chính. Một là tìm một 
câu trả lời hoặc là một giải pháp. Hai là tạo ra một sản phẩm gì đó. Ba 
là tối ưu cho một vấn đề cũ. 

Bước 1: Phân Tích Vấn Ðề
A. Ðịnh Nghĩa Vấn Ðề và Phân Tích MECE
Problem Statement: Ðịnh nghĩa rõ ràng vấn đề.
MECE (Mutually Exclusive, Collectively Exhaustive): Bóc tách 
vấn đề thành các mảnh không trùng lặp và bao quát toàn bộ.
Flow và Sự Phụ Thuộc: Vẽ sơ đồ luồng công việc và sự phụ thuộc 
giữa các mảnh.

B. Thu Thập Thông Tin và Ðào Sâu Phân Tích
Data Collection: Thu thập dữ liệu liên quan đến từng mảnh.
Research: Nghiên cứu các thông tin và kiến thức cần thiết.
Root Cause Analysis: Ðào sâu tìm căn nguyên của vấn đề.
Bottleneck Analysis: Tìm nút thắt của vấn đề.
Bước 2: Xác Ðịnh Lại Vấn Ðề và Mục Tiêu
C. Tuyên Bố Vấn Ðề Sửa Ðổi và Ðặt Mục Tiêu
Revised Problem Statement: Viết lại vấn đề sao cho:
Clear: Rõ ràng.
Measurable: Có thể đo lường được.
Actionable: Có thể hành động.
Focused: Tập trung vào các mục tiêu cụ thể.
Goal Setting: Lên mục tiêu rõ ràng và cụ thể.
Bước 3: Tìm Kiếm Giải Pháp
D. Ðộng Não Giải Pháp
Brainstorming: Tạo ra các ý tưởng mới.
Creative Thinking: Suy nghĩ sáng tạo.
Lateral Thinking: Suy nghĩ theo hướng khác biệt.
Design Thinking: Sử dụng tư duy thiết kế để giải quyết vấn đề.
Bước 4: Phát Triển Phương Pháp
E. Phát Triển Cách Tiếp Cận
Hypothesis: Ðưa ra giả thuyết.
Constraint Analysis: Phân tích các ràng buộc.
Framework & Model Design: Thiết kế khung và mô hình.
Strategy & Methodology: Chiến lược và phương pháp.
Scenario Planning: Lập kế hoạch cho các kịch bản khác nhau.
Risk Assessment: Ðánh giá rủi ro.
Bước 5: Ðánh Giá Phương Pháp
F. Ðánh Giá Phương Pháp
Assessment Criteria: Tiêu chí đánh giá.
Feasibility Analysis: Phân tích tính khả thi.
Pros and Cons: Ðánh giá ưu và nhược điểm.
Feedback Loops: Các vòng phản hồi.
Impact Analysis: Phân tích tác động.
Resource Allocation: Phân bổ tài nguyên.
Prioritization: Ưu tiên các giải pháp.
Benchmarking: So sánh với các tiêu chuẩn.
Validation Testing: Kiểm tra xác nhận.
Quantify and Grading: Ðịnh lượng và phân loại.
Determine Constraints: Xác định các ràng buộc.
Predict Outcomes: Dự đoán kết quả.

Bước 7: Ðánh Giá và Ðiều Chỉnh
Evaluation: Ðánh giá kết quả của giải pháp.

cơ hội mới để tối ưu hóa.
